```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
days = [1,2,3,4,5,6,7]
nop1 = [5,10,15,20,25,30,35]
nop2 = [10,20,30,35,40,50,60]
nop3 = [20,12,30,40,60,80,100]
plt.stackplot(days,nop1,nop2,nop3,colors=['red','blue','pink'],labels=['week1','week2','week3'])
plt.legend()
plt.show()
```
