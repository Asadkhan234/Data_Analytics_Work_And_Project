```python
import matplotlib.pyplot as plt
```

```python
days = ['Day1','Day2','Day3','Day4','Day5','Day6','Day7']
price1 = [100,400,300,500,200,700,600]
price2 = [200,800,400,700,500,300,100]
```

```python
plt.plot(days,price1,marker='+',ls='--',label='Week1 (Prices)')
plt.plot(days,price2,marker='*',ls='--',color='red',label='Week2 (Prices)')
plt.xlabel('DAYS IN A WEEK')
plt.ylabel('PRICES ON EVERY SINGLE DAY')
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
data = pd.read_excel('expense.xlsx')
data
```

```python
grp_by = data.groupby('Category')['Amount'].max()
plt.bar(grp_by.index.astype(str),grp_by.values)
plt.show()
```

```python
#plt.plot(data['Date'],data['Amount'])
color = ['red','green','blue',]
plt.bar(data['Category'].astype(str),data['Amount'],color=color,label='Category')
plt.legend()
plt.show()
```

```python
plt.plot(data['Date'],data['Amount'],marker='*',ls='--',color='red',label='Total Amount Of Each Date')
plt.legend()
plt.show()
```
