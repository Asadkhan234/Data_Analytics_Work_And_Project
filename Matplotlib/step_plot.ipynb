```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = ['Day1','Day2','Da3','Day4','Day5','Day6','Day7']
y =  [20,30,40,50,10,80,90]

plt.step(x,y,where='mid',marker='o',color ='red')
plt.show()
```

```python
d = pd.read_excel("expense.xlsx", engine="openpyxl")  # make sure openpyxl is installed
print(d.head())

group = d.groupby('Category', as_index=False).agg({'Amount':'sum'})
print(group)


plt.step(group.index,group['Amount'],marker='o',where='mid',color='blue',label='Amount')
plt.legend()
plt.title('Cateogry Total Amount')
plt.show()
```

```python
A = d.groupby('Payment Mode',as_index=False)['Amount'].sum()
plt.step(A.index,A['Amount'],marker='*',label='Payment Mode')
plt.legend(loc=4)
plt.show()
```
