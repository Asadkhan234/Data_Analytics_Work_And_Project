```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.DataFrame({
    'Age': [20,23,25,28,29],
    'Salary' : [20000,30000,40000,50000,60000],
    'Country' : ['Us','Uk','Pakistan','China','Iran'],
    'Employees': ['Saad','Sami','Mut','Salman','Za']
})

data
```

```python
c = ['red','blue','green','yellow','black']
plt.bar(data.Employees ,data.Salary,color=c,edgecolor='red')
plt.xlabel('Countries',fontsize=15,color='Blue')
plt.ylabel('Salaries',fontsize=15,color='Blue')
plt.title('Present Countries Salaries',fontsize=15)
plt.show()
```
