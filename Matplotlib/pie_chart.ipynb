```python
import matplotlib.pyplot as plt
```

```python
brands = ['Samsung','Iphone','Oppo','Vevo','Radmi']
sell = [90,40,60,80,30]
ex  = [0,0,0,0,0.1]
c = ['silver','red','green','yellow','blue']
plt.pie(sell,labels=brands,colors = c ,explode=ex,radius=0.9,shadow=True, autopct= '%.2f',rotatelabels=True,startangle=90)
plt.show()
```

```python
import pandas as pd
```

```python
data = pd.read_csv('Customers_Data.csv',encoding='latin')
data
```

```python
data['AnnualIncome'] = data['AnnualIncome'].str.replace('$','')
data['AnnualIncome'] = data['AnnualIncome'].str.replace(',','')
```

```python
data['AnnualIncome'].unique()
```

```python
data['AnnualIncome'] = data['AnnualIncome'].astype(int)
```

```python
grp = data.groupby('EmailAddress')['AnnualIncome'].sum()
```

```python
plt.pie(grp.values,labels=data['FirstName'])
plt.show()
```

```python
plt.pie(data['AnnualIncome'])
plt.show()
```

```python
import pandas as pd
df = pd.read_excel('expense.xlsx')
df
```

```python
import  matplotlib.pyplot as plt
```

```python
plt.plot(df['Payment Mode'].astype(str),df['Amount'])
plt.show()
```

```python
plt.pie(df['Amount'],labels=df['Payment Mode'],shadow=True, autopct= '%.2f')
plt.show()
```

```python
grr = df.groupby('Payment Mode')['Amount'].sum()
plt.pie(grr.values,labels=grr.index,autopct='%.2f',shadow=True,explode=[0,0,0.04])
plt.show()
```

```python
plt.pie(df['Amount'],labels=df['Payment Mode'])
plt.show()
```
