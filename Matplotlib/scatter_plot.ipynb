```python
#Scatter Plot Is Used Where Both x and y is Numeric.
```

```python
import matplotlib.pyplot as plt
import random 
import numpy as np


x = np.random.randint(1,30,50)
print(x)
y = np.random.randint(30,60,50)
print(y)
```

```python
color= np.random.randint(10,100,50)
size = np.random.randint(10,100,50)
plt.scatter(x,y,label='random',cmap='hot',c = color,s = size,marker='*')
plt.colorbar()
plt.legend()
plt.show()
```

```python
import pandas as pd
data = pd.read_csv('student_scores.csv')
data
```

```python
grp = data.groupby('NrSiblings')['MathScore'].count()
print(grp)
plt.scatter(grp.index,grp.values)
plt.show()
plt.scatter(data['NrSiblings'],data['MathScore'])
plt.show()
```

```python
#score = data.groupby(['MathScore','ReadingScore','WritingScore'])['NrSiblings'].count().reset_index()
score = data.groupby('MathScore')['NrSiblings'].count()
plt.scatter(score.index,score.values)
plt.show()
```

```python
s = data.groupby('MathScore')['WritingScore'].count()
plt.plot(s.index, s.values, marker='o')
plt.xticks(rotation=90)
plt.ylabel("Count of WritingScore")
plt.title("Counts by Math & Reading Score")
plt.show()
```
