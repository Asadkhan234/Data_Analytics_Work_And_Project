```python
# Working On data set :
data = pd.read_csv('Customers_Data.csv',encoding='latin')
data
```

```python
data['AnnualIncome'] = data['AnnualIncome'].str.replace('$','')
data['AnnualIncome'] = data['AnnualIncome'].str.replace(',','')
data['AnnualIncome'] = data['AnnualIncome'].astype(int)

data['AnnualIncome']
```

```python
plot = [data['AnnualIncome'],data['TotalChildren']]
plt.boxplot(plot)
plt.show()
```

```python
male = data[data['Gender']=='M']['AnnualIncome']
female = data[data['Gender']=='F']['AnnualIncome']

plt.boxplot([male, female], labels=['Male', 'Female'])
plt.title("Annual Income Distribution by Gender")
plt.ylabel("Annual Income")
plt.show()
```

```python
gen = data.groupby('Gender')['AnnualIncome'].count()
print(gen)
plt.boxplot(gen)
plt.show()
```

```python
mari_sing = data.groupby('MaritalStatus')['AnnualIncome'].count()
mari_sing
```

```python
plt.boxplot(mari_sing)
plt.show()
```

```python
maried = data[data['MaritalStatus']=='M']['AnnualIncome']
single = data[data['MaritalStatus']=='S']['AnnualIncome']
mari_single = [maried,single]
plt.boxplot(mari_single,labels=['Maried','Single'])
plt.show()
```

```python
count = data.groupby('TotalChildren')['AnnualIncome'].count()
print(count)
plt.boxplot(count)
plt.show()
```

```python
Edu_count = data[data['EducationLevel']=='Bachelors']['AnnualIncome']
Edu_sum = data[data['EducationLevel']=='Graduate Degree']['AnnualIncome']
Edu = [Edu_count,Edu_sum]
```

```python
plt.boxplot(Edu,labels=['Bachelors','Graduate Degree'])
plt.show()
```
