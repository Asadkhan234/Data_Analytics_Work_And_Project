# Using Of group by :
- we use it to get data from the columns .

```python
import pandas as pd
data = pd.read_excel('ESD (3).xlsx')
data.head(10)
```

```python
# using of group by :
gp = data.groupby('Department')['Gender'].count()
gp
```

```python
gp = data.groupby(['Department','Gender']).agg({'EEID': 'count'})
gp
```

```python
gp.count()
```

```python
data
```

```python
data.groupby('Full Name')['City'].value_counts()
```

```python
data.groupby('Country').agg({'Age':'mean'})
```

```python
data.groupby(['Country','City']).agg({'Annual Salary':'max'})
```

```python
one = data.groupby(['Full Name','EEID']).agg({'Hire Date':'mean'})
```

```python
data.groupby(['Full Name','Gender']).agg({'Ethnicity':'count'})
```

```python
data.groupby(['Full Name','Gender']).agg({'Ethnicity':'first'})
```

```python
data.groupby(['Full Name','Gender']).agg({'Ethnicity':lambda x : ','.join(x.unique())})
```
