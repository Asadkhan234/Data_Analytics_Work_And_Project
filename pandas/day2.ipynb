```python
data['GetBonus %'] = np.where(data['Bonus %'] == 0, 'No Bonus', 'Bonus')
data['GetBonus %'].value_counts()
```

```python
data.loc[data['Age'] <= 40, 'New_Ages'] = 'Young'
data.loc[data['Age'] > 40,'New_Ages'] = 'Old'
```
