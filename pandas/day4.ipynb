# Merge - Joins - Concatinate :

# Merge :

```python
import pandas as pd

# First DataFrame: employee details
df1 = pd.DataFrame({
    'EEID': [101, 102, 103, 104],
    'Full Name': ['Ali Khan', 'Sara Ahmed', 'John Doe', 'Maryam Ali'],
    'Department': ['HR', 'IT', 'Finance', 'IT']
})

# Second DataFrame: extra employee info
df2 = pd.DataFrame({
    'EEID': [101, 102, 103, 105],
    'Gender': ['Male', 'Female', 'Male', 'Female'],
    'Ethnicity': ['Asian', 'Arab', 'Caucasian', 'Asian']
})

print("DataFrame 1:\n", df1)
print("\nDataFrame 2:\n", df2)
```

```python
pd.merge(df1,df2,on='EEID',how='inner')
```

```python
pd.merge(df1,df2,on='EEID',how='outer')
```

```python
pd.merge(df1,df2,on='EEID',how='left')
```

```python
pd.merge(df1,df2,on='EEID',how='right')
```

# Contatinate

```python

import pandas as pd

# First DataFrame: employee details
df1 = pd.DataFrame({
    'EEID': [101, 102, 103, 104],
    'Full Name': ['Ali Khan', 'Sara Ahmed', 'John Doe', 'Maryam Ali'],
    'Department': ['HR', 'IT', 'Finance', 'IT']
})

# Second DataFrame: extra employee info
df2 = pd.DataFrame({
    'EEID1': [101, 102, 103, 105],
    'Gender': ['Male', 'Female', 'Male', 'Female'],
    'Ethnicity': ['Asian', 'Arab', 'Caucasian', 'Asian']
})

print("DataFrame 1:\n", df1)
print("\nDataFrame 2:\n", df2)
```

```python
pd.concat([df1,df2],axis=1).drop(columns=['EEID1'],axis=1)
```

```python
pd.concat([df1,df2],axis=0).drop(columns=['EEID1'],axis=0)
```

# Compare : 
- Compare Two Data Frames
- Cheack What Changes Comes In The New One Data Frame .

```python
df1 = pd.DataFrame({
    'Fruits':['Mango','Apple','Bannana','Papaya'],
    'Prices':[100,150,50,35],
    'Quantity':[15,10,10,3]
})

df2 = df1.copy()
```

```python
df2['Prices'] = df2['Prices']=[102,175,50,30]
```

```python
df2['Quantity']= df2['Quantity']=[12,15,10,5]
```

```python
df1
```

```python
df2
```

```python
df1.compare(df2)
```

```python
df1.compare(df2,keep_equal=True)
```

```python
df1.compare(df2,keep_equal=False)
```

```python
df1.compare(df2,keep_shape=True)
```

```python
df1.compare(df2,keep_equal=False)
```

# Pevioting : 
- It Os Working On three Parameters
- Index
- column
- values

```python
data = pd.DataFrame({
    'Keys': ['K1','K2','K3','K1'],
    'Names' : ['Asad','Saad','Sami','Mutahir'],
    'Salary' : [20000,40000,50000,30000]
})


data.loc[:,'Proffession'] = ['Worker','Doc','Sport','IT']
data
```

```python
data.pivot(index = 'Keys',columns = 'Names',values =[ 'Salary','Proffession'])
```

```python
d= pd.read_csv('employees.csv')
d
```

```python
d.loc[:,'Full Name'] = d['First name'].str.capitalize() +' '+ d['Last Name'].str.capitalize()
```

```python
d.pivot(index = 'EEID' , columns = 'Full Name' , values = ['Salary','First name'])
```
